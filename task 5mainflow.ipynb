```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Setting maximum number of columns and rows to display all fields in the output
pd.set_option('display.max_columns', 110)
pd.set_option('display.max_rows', 1100)
```

```python
#reading the data
data=pd.read_csv(r'/content/heart.csv')
```

```python
#Understanding the data
data.info()
```

```python
#Preview the data
data.head()
```

```python
#Checking for null values
data.isna().sum()
```

```python
#Using ggplot style for better visualization
plt.style.use('ggplot')
```

```python
#Statistical summary of the data
data.describe()
```

```python
#Histograms of all the columns
data.hist(figsize=(13,13), bins=40, grid=False, color='skyblue')
plt.show()
```

```python
#Correalation matrix of the data
plt.figure(figsize=(14,6))
correlation=data[data.columns].corr()
sns.heatmap(correlation,annot=True,cmap='Pastel1', fmt=".2f")
plt.show()
```

```python
pd.crosstab(data.age.value_counts(), data.sex)
```

```python
data.age.mean()
```

```python
sns.displot(data=data, x='age', hue='sex', palette=['#1f77b4', '#ff7f0e'], bins=30, height=3, aspect=2)
plt.title('Age Distribution')
plt.xlabel('Age (0=Female, 1=Male)')
plt.show()
```

```python
data.groupby(['sex','age'])['age'].count()
```

```python
sns.jointplot(data=data, x='age',y='chol',hue='sex', kind='hist',height=5)
plt.ylabel('Cholestrol found')
plt.xlabel('Age of the patient (0=Male, 1=Female)')
plt.show()
```

```python
data[data['target']==1]['target'].value_counts()
```

```python
pd.crosstab(data.target, data.chol)
```

```python
plt.figure(figsize=(4,4))
sns.histplot(data=data,x='chol', hue='target',kde=True)
plt.title('Cholestrol levels & Heart Disease')
plt.xlabel('Cholestrol Levels')
legend = plt.legend(title='Heart Disease', labels=['Yes', 'No'])
plt.show()
```

```python
sns.scatterplot(x='thalach',y='age',data=data,hue='chol')
plt.title('Cholestrol Levels & Thalach as per Age')
plt.xlabel('Thalach')
plt.ylabel('Age')
plt.show()
```

```python
pd.crosstab(data.target, data.cp).plot(kind='barh', figsize=(10,5), color=['#859644', '#97e3fd', '#d34986', '#d1d349'])
plt.title('Heart Disease Frequency as per CP')
plt.yticks([0, 1], ['Disease', 'No Disease'])
plt.xlabel('CP')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=data,x='slope', hue='target')
plt.xticks([0, 1, 2], ['Unsloping','Unsloping Slope','Downsloping'])
legend = plt.legend(title='Heart Disease', labels=['Yes', 'No'])
plt.show()
```
